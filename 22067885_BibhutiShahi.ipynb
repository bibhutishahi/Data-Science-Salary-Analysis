```python
#Write a python program that loads data into pandas dataframe.

import pandas as pd

#Loading the data into the pandas.

dataSet = pd.read_csv("SalarySheet.csv")
df = pd.DataFrame(dataSet)
df
```

```python
#Write a python program to remove the unnecessary columns i.e., salary and salary currency.

unnecessaryColumns = ['salary', 'salary_currency']
df= df.drop(columns = unnecessaryColumns)
df
```

```python
#Write a python program to remove the NaN missing values from updated DataFrame.

df = df.dropna()
df
```

```python
#Write a python program to check duplicates value in the dataframe.

duplicatevalues = df[df.duplicated()]
duplicatevalues
```

```python
#Write a python program to see the unique values from all the columns in the dataframe

#Creating an empty dictionary to store thevalues

unique_values_dictionary = {}

#Iterating through each column into the Dataframe

for column in df.columns:
    unique_values_dictionary[column] = df[column].unique()
    
#Displaying unique values of each of the column

for column, values in unique_values_dictionary.items():
    print(f"The unique values in the column '{column}':")
    print(values)
    print()
```

```python
#Renaming the post to senior level

df['experience_level'] = df['experience_level'].replace("SE" , "Senior Level")
df
```

```python
#Renaming MI level to Medium level

df['experience_level'] = df['experience_level'].replace("MI" ," Medium Level")
df
```

```python
#Renaming EN to Entry Level

df['experience_level'] = df['experience_level'].replace("EN", "Entry Level")
df
```

```python
#Renaming EX level to Execute level

df['experience_level'] = df['experience_level'].replace("EX", "Execute Level")
execute_level = df[df['experience_level'] == 'Execute Level']
execute_level
```

```python
#write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable

# Asking user for column

chosenvariable = input("Enter the name of the variable you want to analyze: ") 

# Verifying if the chosen variable exists in the df DataFrame 

if chosenvariable in df.columns: 
    # Describing method to get all the statistics 
    summarystats = df[chosenvariable].describe() 
    
    # Calculating skewness and kurtosis 
    skewness = df[chosenvariable].skew() 
    kurtosis = df[chosenvariable].kurtosis() 
    
    # Displaying the summary statistics 
    print ("Summary Statistics:") 
    print (summarystats)
    print("\nAdditional Statistics:") 
    print(f"Skewness: {skewness}") 
    print(f"Kurtosis: {kurtosis}") 
else: 
    print(f"Error: The variable '{chosenvariable}' does not exist in the DataFrame.")
```

```python
# Select only numeric columns as pandas cannot calculate correlation for strings 

numericdf = df.select_dtypes (include=['number']) 

#Calculating the correlation of the numeric values 

correlation_matrix = numericdf.corr()

# Displaying the correlation matrix 
print("Correlation Matrix:") 
print (correlation_matrix)
```

```python
#Finding the top 15 jobs 
#count - count unique values 
#head(15) - to display the top 15 details/ 
#tolist() - converted to List for better display and easier manipulation 

Top_Fifteen_Jobs = df['job_title'].value_counts().head(15).index.tolist() 
Top_Fifteen_Jobs
```

```python
import matplotlib.pyplot as plt 
import numpy as np 

#Finding the top 15 jobs 
#count - count unique values 
#head(15) - to display the top 15 details/ 
#tolist) - converted to List for better display and easier manipulation 

Top_Fifteen_Jobs = df['job_title'].value_counts().head(15).index.tolist() 
Top_Fifteen_Jobs 
y= Top_Fifteen_Jobs
x= np.arange(1,16) 
plt.bar(x,y) 
plt.xlabel("Count") 
plt. ylabel("Top jobs")
```

```python
#Write a program to find out the job with highest salary with bar graph 

highest_salary = df.groupby('job_title')['salary_in_usd'].max().sort_values(ascending=False).head(20)

#Plotting the bar graph
plt.figure(figsize=(15, 10))
highest_salary.plot(kind='bar')
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')
plt.title('Top Salary by Job Title')
plt.xticks(rotation=90)
plt.show()
```
